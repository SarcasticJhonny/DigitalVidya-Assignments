We get stocks data using package pandas_datareader. You will have to first install pandas_datareader using either of the commands given below (pip or pip3 or conda). We have selected Infy, HDFC Bank, Tata Motors and Wipro which are listed on NYSE. Once we get the data, we will do some aggregations using pandas functions and datetime funcionalities.

```python
#conda install -c https://conda.anaconda.org/anaconda pandas-datareader
#conda install -c anaconda pandas-datareader=0.4.0
#pip3 install pandas_datareader
#pip install pandas_datareader

import pandas as pd
import numpy as np
import pandas_datareader.data as web
import datetime as dt

start = dt.datetime(2014, 1, 1)

end = dt.datetime.today()

stocks = ['INFY', 'HDB', 'TTM', 'WIT']

df = web.DataReader(stocks, 'google', start, end)
df.items
```

The above df is of the type panel, which is a 3-D array. You can create a dataframe from that by calling df[index_name] and assiging it to the desired dataframe name.

#### 1. Create a dataframe called open_prices using open prices of the listed stocks.

```python
open_prices = df['Open']
open_prices.head()
```

#### 2. Use the dataframe open_prices created in qn 1 to create open_yearly with mean open prices every year. 

```python
open_yearly = open_prices.groupby(open_prices.index.year).mean()
open_yearly
```

#### 3. Group the open_prices to get the mean prices per week per year. The resulting dataframe open_weekly should have mean price data for week 1 to 52 for year 2015, 2016 and 2017. 
Hint: You may have to groupby open_prices.index.week and open_prices.index.year to ensure that you get data per week per year.

```python
open_weekly = open_prices.groupby([open_prices.index.year, open_prices.index.week]).mean()
open_weekly
```

#### 4. Create a dataframe vol with Volume of all the four stocks traded during the given time period.

```python
vol = df['Close']
vol.head()
```

#### 5. Create dataframe volume_yearly with total volume per year.

```python
volume_yearly = vol.groupby(vol.index.year).sum()
```

#### 6. Create dataframe volume_weekly with total volume per week per year.

```python
volume_weekly = vol.groupby([vol.index.year, vol.index.week]).sum()
volume_weekly
```

#### 7. Create a dataframe close_prices using df['Close']

```python
close_prices = df['Close']
close_prices.head()
```

#### 8. Use the dataframe close_prices to find the min close price of the stocks in each month per year.

```python
close_prices.groupby([close_prices.index.year, close_prices.index.month]).min()
```

#### 9. You can extract the price of an individual stock using the following lines of code. Find the number of days the close price was greater than open price

```python
Amazon = web.DataReader('AMZN', 'google', start, end)
Amazon.head()
```

```python
Amazon.loc[Amazon['Close'] > Amazon['Open'], ['Close', 'Open']].index.day.value_counts()
```

#### 10. Find the maximum close price of Amazon and the date on which it was maximum

```python
Amazon = Amazon.reset_index()
print('The maximum Close price was:',format(Amazon.loc[Amazon['Close'].argmax(),['Date','Close']][1]),'\n','On:',format(Amazon.loc[Amazon['Close'].argmax(),['Date','Close']][0]))
```
